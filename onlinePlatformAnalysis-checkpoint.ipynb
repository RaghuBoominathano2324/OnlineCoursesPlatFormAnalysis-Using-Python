#Enrollment Analysis

1.	Which course categories have the highest student enrollments on each platform?
Outcome: Identify the most popular categories among learners.
2.	Do certain categories consistently have higher enrollments across both platforms?
Outcome: Highlight universal interest areas for learners.




```python
import matplotlib.pyplot as plt  # Make sure this import is included
import seaborn as sns

# Grouping by Platform and Course Category, summing up enrollments
enrollment_data = data.groupby(['Platform', 'Course Category'])['Number of Students Enrolled'].sum().reset_index()

# Sorting data for clear visualization
enrollment_data = enrollment_data.sort_values(by='Number of Students Enrolled', ascending=False)

# Setting the style and context for a more professional look
sns.set_style('whitegrid')
sns.set_context('talk', font_scale=1.3)

# Creating the plot with larger figure size
plt.figure(figsize=(18, 10))

# Plotting with a more vibrant color palette and adding 'dodge' to separate bars
bar_plot = sns.barplot(
    x='Number of Students Enrolled',
    y='Course Category',
    hue='Platform',
    data=enrollment_data,
    palette='magma',
    dodge=True
)

# Adding titles and labels with better alignment and font
plt.title('Course Enrollment Analysis by Platform and Category', fontsize=20, weight='bold', pad=20)
plt.xlabel('Number of Students Enrolled', fontsize=16, labelpad=15)
plt.ylabel('Course Category', fontsize=16, labelpad=15)

# Customizing the legend with a more refined style
plt.legend(title='Platform', title_fontsize=14, fontsize=12, loc='upper right', bbox_to_anchor=(1.1, 1))

# Adding data labels on top of bars for clarity
for p in bar_plot.patches:
    bar_plot.annotate(
        format(p.get_width(), ',.0f'),  # Adding commas for readability
        (p.get_width(), p.get_y() + p.get_height() / 2),
        xytext=(5, 0),
        textcoords='offset points',
        ha='left',
        va='center',
        fontsize=12,
        color='black',
        fontweight='bold'
    )

# Adding gridlines for better readability, only on x-axis
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Tight layout for better spacing
plt.tight_layout()

# Display the plot
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting Review Count vs. Rating
plt.figure(figsize=(14, 8))

# Scatter plot with bubble size representing the number of students enrolled
scatter_plot = sns.scatterplot(
    data=data,
    x='Review Count',
    y='Rating',
    size='Number of Students Enrolled',  # Bubble size
    hue='Platform',  # Color by platform
    palette='coolwarm',  # Better contrasting color palette
    sizes=(100, 2000),  # Adjust size range for better clarity
    alpha=0.7  # Set transparency for overlapping bubbles
)

# Add a regression trendline (linear fit)
sns.regplot(
    data=data,
    x='Review Count',
    y='Rating',
    scatter=False,  # Only show the line
    color='red',
    line_kws={'linestyle': '--'},  # Line style for the regression line
    label='Trendline'
)

# Highlighting the top course (max enrolled students)
top_course = data.loc[data['Number of Students Enrolled'].idxmax()]
plt.text(
    top_course['Review Count'], 
    top_course['Rating'] + 0.1, 
    f"Top Course\n({int(top_course['Number of Students Enrolled']):,} enrolled)", 
    color='darkblue',
    fontsize=12,
    weight='bold'
)

# Adjusting legend for platform and bubble size
legend = plt.legend(
    title='Legend',
    bbox_to_anchor=(1.05, 1),
    loc='upper left',
    fontsize=12,
    labels=['Trendline', 'Coursera', 'Udemy', 'Bubble Size: Students Enrolled']
)
legend.get_title().set_fontsize(14)

# Title and axis labels
plt.title('Enhanced Review Count vs. Rating (Bubble Size: Number of Students Enrolled)', fontsize=18, weight='bold')
plt.xlabel('Review Count', fontsize=14)
plt.ylabel('Rating', fontsize=14)

# Adding subtle gridlines for better readability
plt.grid(visible=True, which='both', axis='both', linestyle='--', alpha=0.5)

# Tight layout to ensure proper display
plt.tight_layout()

# Display the plot
plt.show()
```

#Most Popular Courses Word Cloud
6. What is the most popular course keyword based on student enrollments?
	Outcome: Identify which keywords (e.g., "Python," "Data Science") are linked with the highest enrollments.
Visualization: Word cloud or bar chart for the most frequent Course Keywords.



```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import matplotlib.patches as mpatches  # For custom legend

# Load the dataset
data = pd.read_excel('Cleaned_Courses_Dataset_Updated_1.xlsx')

# Extract keywords from the 'Course Name' column
stopwords = set(['the', 'and', 'of', 'in', 'to', 'for', 'on', 'with', 'from', 'by', 'a', 'an', 'is', 'it'])
data['Keywords'] = data['Course Name'].str.lower().str.split()

# Flatten the list of keywords with enrollments
keywords_with_enrollments = []
for i, row in data.iterrows():
    keywords = [word for word in row['Keywords'] if word not in stopwords]
    for keyword in keywords:
        keywords_with_enrollments.append((keyword, row['Number of Students Enrolled']))

# Aggregate enrollments by keyword
keyword_enrollments = pd.DataFrame(keywords_with_enrollments, columns=['Keyword', 'Enrollments'])
keyword_enrollments = keyword_enrollments.groupby('Keyword')['Enrollments'].sum().reset_index()
keyword_enrollments = keyword_enrollments.sort_values(by='Enrollments', ascending=False)

# Top 20 keywords
top_keywords = keyword_enrollments.head(20)

# 1. Word Cloud (Improved Color & Style)
plt.figure(figsize=(12, 8))
wordcloud = WordCloud(
    width=800, height=400, 
    background_color='black',  # Change background for contrast
    colormap='plasma',  # Use a vibrant color palette
    contour_width=1,  # Slight contour for added depth
    contour_color='gold',  # Gold contour for better contrast and aesthetics
    max_words=200,  # Limit number of words to improve legibility
    min_font_size=20,  # Set minimum font size to improve clarity for smaller words
).generate_from_frequencies(dict(zip(keyword_enrollments['Keyword'], keyword_enrollments['Enrollments'])))

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Popular Course Keywords by Enrollments', fontsize=18, weight='bold', color='white')
plt.tight_layout()
plt.show()

# 2. Bar Chart (Improved Colors, Gridlines & Custom Legend)
plt.figure(figsize=(12, 8))
sns.set_theme(style="whitegrid")

# Create the bar plot
colors = sns.color_palette("magma", n_colors=20)  # Generate distinct colors for bars
sns.barplot(
    x='Enrollments', 
    y='Keyword', 
    data=top_keywords, 
    ci=None, 
    palette=colors  # Apply the color palette to the bars
)

# Custom legend
legend_labels = [f"{keyword}" for keyword in top_keywords['Keyword']]
legend_patches = [mpatches.Patch(color=colors[i], label=label) for i, label in enumerate(legend_labels)]

# Add legend to the chart
plt.legend(
    handles=legend_patches,
    bbox_to_anchor=(1.05, 1),  # Position legend outside the chart
    loc='upper left', 
    title='Keywords', 
    fontsize=10,
    title_fontsize=12
)

# Add gridlines for better readability
plt.grid(axis='x', linestyle='--', alpha=0.6)

# Titles and labels
plt.title('Top 20 Course Keywords by Total Enrollments', fontsize=18, weight='bold', color='darkslategray')
plt.xlabel('Total Enrollments', fontsize=14, color='darkslategray')
plt.ylabel('Keyword', fontsize=14, color='darkslategray')

# Improve the layout
plt.tight_layout()
plt.show()
```

## Price and Rating Analysis
Free vs Paid Enrollment
7.Which platform sees higher student interest in free courses versus paid ones?
Outcome:
Compare enrollments for free vs. paid courses for each platform.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'data' is your DataFrame containing the course information

# Group the data by Platform and 'IF Free/Not' (assuming '1' means free and '0' means paid)
enrollment_data = data.groupby(['Platform', 'IF Free/Not'])['Number of Students Enrolled'].sum().unstack()

# Set a custom color palette
colors = ['#4C72B0', '#DD8452']  # Blue for Paid, Salmon for Free

# Create the plot
plt.figure(figsize=(12, 8))
bar_width = 0.35
enrollment_data.plot(kind='bar', figsize=(12, 8), color=colors, width=bar_width)

# Titles and labels
plt.title('Free vs. Paid Enrollment by Platform', fontsize=18, weight='bold')
plt.xlabel('Platform', fontsize=14, labelpad=10)
plt.ylabel('Number of Students Enrolled', fontsize=14, labelpad=10)
plt.xticks(rotation=0, fontsize=12)  # Keep platform names horizontal
plt.yticks(fontsize=12)
plt.legend(['Paid', 'Free'], title='Course Type', fontsize=12, title_fontsize=14, loc='upper right')

# Add gridlines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adding numerical labels on each bar
for i, platform in enumerate(enrollment_data.index):
    for j, course_type in enumerate(enrollment_data.columns):
        value = enrollment_data.loc[platform, course_type]
        plt.text(
            x=i + (j - 0.5) * bar_width,  # Centering text on the bar
            y=value + (0.02 * max(enrollment_data.max())),  # Slightly above the bar
            s=f"{int(value):,}",  # Formatting numbers with commas
            ha='center', 
            fontsize=10,
            weight='bold',
            color='black'
        )

# Tight layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Generate a summary of enrollments for written output
summary = []
for platform in enrollment_data.index:
    for course_type in enrollment_data.columns:
        value = enrollment_data.loc[platform, course_type]
        course_type_label = "Free" if course_type == 1 else "Paid"
        summary.append(f"{platform}: {int(value):,} students enrolled in {course_type_label} courses.")

# Print the summary
print("\nEnrollment Summary by Platform and Course Type:\n")
for line in summary:
    print(line)
```

# Price and Rating Analysis
Rating By Price Catogery
8.	Do higher-priced courses tend to receive better ratings on average?

9.	Which price category shows the widest range of ratings for courses?


11. Which price category shows the widest range of ratings for courses?


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
data = pd.read_excel('Cleaned_Courses_Dataset_Updated_1.xlsx')

# Create a new column for Price Category
def price_category(price):
    if price == 0:
        return 'Free'
    elif price <= 50:
        return 'Low'
    elif price <= 150:
        return 'Medium'
    else:
        return 'High'

data['Price Category'] = data['Price(in $)'].apply(price_category)

# Set a global theme for plots
sns.set_theme(style="whitegrid", font_scale=1.2)

# 1. Bar Chart: Average Ratings by Price Category
average_ratings = data.groupby('Price Category')['Rating'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(
    x='Price Category', 
    y='Rating', 
    data=average_ratings, 
    palette='coolwarm',  # Vibrant color palette
    edgecolor='black'    # Add edge color for better definition
)

# Adding labels on top of the bars
for i, row in average_ratings.iterrows():
    plt.text(
        x=i, 
        y=row['Rating'] + 0.02, 
        s=f"{row['Rating']:.2f}", 
        ha='center', 
        fontsize=11, 
        weight='bold'
    )

# Title and axis labels
plt.title('Average Ratings by Price Category', fontsize=18, weight='bold')
plt.xlabel('Price Category', fontsize=14, labelpad=10)
plt.ylabel('Average Rating', fontsize=14, labelpad=10)
plt.ylim(0, 5)  # Set consistent y-axis range for ratings

# Adding a legend for clarification
handles = [plt.Rectangle((0, 0), 1, 1, color=color) for color in sns.color_palette("coolwarm", 4)]
labels = ['Free', 'Low', 'Medium', 'High']
plt.legend(handles, labels, title='Price Categories', loc='upper right', fontsize=12, title_fontsize=13)

plt.tight_layout()
plt.show()

# 2. Box Plot: Distribution of Ratings by Price Category
plt.figure(figsize=(12, 8))
sns.boxplot(
    x='Price Category', 
    y='Rating', 
    data=data, 
    palette='Spectral',  # Use a distinct color palette
    width=0.6,           # Adjust box width for better spacing
    linewidth=1.5        # Make boxplot lines more prominent
)

# Highlight the median with custom formatting
plt.axhline(y=4, color='gray', linestyle='--', linewidth=1, alpha=0.7)

# Title and axis labels
plt.title('Distribution of Ratings by Price Category', fontsize=18, weight='bold')
plt.xlabel('Price Category', fontsize=14, labelpad=10)
plt.ylabel('Ratings', fontsize=14, labelpad=10)

# Add a legend for boxplot
legend_elements = [
    plt.Line2D([0], [0], color='gray', lw=2, linestyle='--', label='Benchmark Median (4.0)')
]
plt.legend(handles=legend_elements, loc='upper right', fontsize=12)

# Add custom y-axis ticks
plt.yticks(range(0, 6), fontsize=12)

# Improve layout
plt.tight_layout()
plt.show()
```

##Weighted Average Rating based on platform

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_excel("Cleaned_Courses_Dataset_Updated_1.xlsx")

# Handle missing or invalid values in the 'Rating' and 'Review Count' columns
data = data.dropna(subset=['Rating', 'Review Count'])  # Remove rows with missing ratings or review counts
data = data[data['Review Count'] > 0]  # Filter out courses with zero review count

# Calculate weighted average rating for each platform
data['Weighted Rating'] = data['Rating'] * data['Review Count']

# Group by platform and calculate the weighted average rating
platform_avg_rating = data.groupby('Platform').apply(
    lambda x: x['Weighted Rating'].sum() / x['Review Count'].sum()
).reset_index(name='Weighted Average Rating')

# Plotting the weighted average ratings for both platforms
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")  # Set a theme for better gridlines
sns.barplot(
    x='Platform', 
    y='Weighted Average Rating', 
    data=platform_avg_rating,
    palette='Blues_d',  # Use a different palette for more appeal
)

# Add value labels on top of each bar
for i in range(len(platform_avg_rating)):
    plt.text(
        i, platform_avg_rating['Weighted Average Rating'][i] + 0.05,  # Position above the bars
        round(platform_avg_rating['Weighted Average Rating'][i], 2),  # Rounded value to 2 decimal places
        ha='center', va='bottom', fontsize=12, color='black'
    )

# Add titles and labels with enhanced font size and styles
plt.title('Comparison of Weighted Average Ratings by Platform', fontsize=18, weight='bold')
plt.xlabel('Platform', fontsize=14)
plt.ylabel('Weighted Average Rating', fontsize=14)

# Improve layout for better presentation
plt.tight_layout()

# Display the plot
plt.show()

# Print a summary of the weighted average ratings per platform
for i, row in platform_avg_rating.iterrows():
    print(f"Platform: {row['Platform']} - Weighted Average Rating: {row['Weighted Average Rating']:.2f}")
```
